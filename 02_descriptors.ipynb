```python
import deepchem as dc
from rdkit import Chem
import pandas as pd
```

- RDKIT descriptors

```python
mol = Chem.MolFromSmiles('CCC')
featurizer = dc.feat.RDKitDescriptors()
rdkit_mol = featurizer.featurize(mol)
desc_rdkit = [i[0] for i in featurizer.descList]
```

```python
pd.DataFrame(rdkit_mol, columns=desc_rdkit)
```

- circular fingerprint

```python
mol = Chem.MolFromSmiles('CCC')
featurizer = dc.feat.CircularFingerprint(size=1024)
desc_mol = featurizer.featurize(mol)
desc_circular = [f"circular_{i}" for i in range(desc_mol.shape[1])]
```

```python
pd.DataFrame(desc_mol, columns=desc_circular)
```

- MACCS

```python
mol = Chem.MolFromSmiles('CCC')
featurizer = dc.feat.MACCSKeysFingerprint()
desc_mol = featurizer.featurize(mol)
desc_maccs = [f"maccs_{i}" for i in range(desc_mol.shape[1])]
```

```python
pd.DataFrame(desc_mol, columns=desc_maccs)
```

- mordred

```python
mol = Chem.MolFromSmiles('CCC')
featurizer = dc.feat.MordredDescriptors(ignore_3D=False)
desc_mol = featurizer.featurize(mol)
desc_mordred = [f"mordred_{i}" for i in range(desc_mol.shape[1])]
```

```python
pd.DataFrame(desc_mol, columns=desc_mordred)
```

# Mordred

```python
from mordred import Calculator, descriptors
```

```python
calc = Calculator(descriptors)

mol = Chem.MolFromSmiles('CCC')
desc_mol = calc(mol)

desc_mordred = desc_mol.asdict().keys()
desc_mol = desc_mol.asdict().values()
```

```python
pd.DataFrame([desc_mol], columns=desc_mordred)
```
